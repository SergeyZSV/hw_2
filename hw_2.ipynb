### 4.3 Как много домов коснулся ремонт?

```python
renovation = []
for i in df['yr_renovated']:
    if i != 0:
        renovation.append(1)
    else:
        renovation.append(0)
df.insert(15, 'renovation_status', renovation)
```

### 4.4 Насколько в среднем стоимость домов с оценкой grade  выше 10 отличается от стоимости домов с оценкой grade меньше 4?

## Задача 5

### 5.1 Выберите дом клиенту
Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

```python
variants = int(df[(df['waterfront'] == 1) & (df['bathrooms'] >= 3) & (df['sqft_basement'] > 0)][['id']].count())
```

```python
print(f'Количество вариантов для клиента: {variants}')
```

### 5.2 Выберите дом клиенту
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
df_variants = df[
    ((df['view'] == df['view'].max()) | (df['waterfront'] == 1))
    & (df['grade'] == df['grade'].max())
    & (df['yr_built'] >= 1980)
]
```

```python
min_price = int(df_variants['price'].min())
max_price = int(df_variants['price'].max())
price_spread = f'{min_price} - {max_price}'
```

```python
print(f'Диапазон цен: {price_spread}')
```

### 5.3 Выберите дом клиенту
Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

```python
df_variants = df[
    (df['sqft_basement'] == 0)
    & (df['floors'] == 2)
    & (df['price'] < 150000)
]
```

```python
cond = round(df_variants['condition'].mean(), 2)
```

```python
print(f'Среднняя оценка домов по состоянию: {cond}')
```
